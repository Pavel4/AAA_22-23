```python
import matplotlib.pyplot as plt
import numpy as np
import datetime

f, ax = plt.subplots(2, 2)
f.suptitle('Plot examples', fontsize=20)
f.set_figheight(8)
f.set_figwidth(12)

# Scatter plot
cov_ = [[1, 0.7],[0.7, 1]]
mean_ = [20, 10]
x, y = np.random.multivariate_normal(mean_, cov_, size=1000).T
ax[0, 0].set_title('Scatter plot example')
ax[0, 0].scatter(x, y, edgecolors = 'black')

# Hist plot
x = np.random.geometric(p=0.1, size=3000)
ax[0, 1].set_title('Hist plot example')
ax[0, 1].hist(x, bins=40, color='red')


# Line plot
x = np.linspace(0, 6, 1000)
ax[1, 0].set_title('Line plot example')
ax[1, 0].plot(x, np.log(x), color='black', label='log(x)')
ax[1, 0].plot(x, np.log(2 * x), color='red', label='log(2x)')
ax[1, 0].legend(loc='upper left')

# Bar plot
base = datetime.datetime(2021, 1, 1)
dates = [str(i).split()[0] for i in np.array([base + datetime.timedelta(days=i) for i in range(14)])]
x = np.random.normal(5, size=14)
ax[1, 1].set_title('Bar plot example')
ax[1, 1].bar(dates, x)
ax[1, 1].tick_params(axis='x', labelrotation=90)

f.set_tight_layout(True)
f.savefig('plot_examples.png', facecolor='white')
```
